# Complejidad Algorítmica: Comparación Visual

¿Cómo crece el tiempo de ejecución cuando aumentamos el tamaño de los datos?
Vamos a comparar tres complejidades clásicas:
- $O(n)$ — lineal
- $O(n \log n)$ — linealítica
- $O(n^2)$ — cuadrática

```python
import numpy as np
import matplotlib.pyplot as plt
import time

def medir_tiempo(func, datos):
    inicio = time.perf_counter()
    func(datos)
    return time.perf_counter() - inicio

def busqueda_lineal(lista):
    objetivo = lista[-1]
    for x in lista:
        if x == objetivo:
            return x

def ordenar(lista):
    return sorted(lista)

def burbuja(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]

tamanos = [100, 500, 1000, 2000, 3000, 5000]
t_on, t_onlogn, t_on2 = [], [], []

for n in tamanos:
    datos = list(range(n, 0, -1))
    t_on.append(medir_tiempo(busqueda_lineal, datos[:]))
    t_onlogn.append(medir_tiempo(ordenar, datos[:]))
    t_on2.append(medir_tiempo(burbuja, datos[:]))

plt.figure(figsize=(8, 5))
plt.plot(tamanos, t_on, 'o-', label='O(n) — Búsqueda lineal')
plt.plot(tamanos, t_onlogn, 's-', label='O(n log n) — Ordenación')
plt.plot(tamanos, t_on2, '^-', label='O(n²) — Burbuja')
plt.xlabel('Tamaño de entrada (n)')
plt.ylabel('Tiempo (s)')
plt.title('Comparación de complejidades algorítmicas')
plt.legend()
plt.grid(True)
plt.show()
```

## Resultado esperado

La gráfica muestra claramente cómo $O(n^2)$ crece mucho más rápido que $O(n)$ y $O(n \log n)$.
Esto explica por qué elegir el algoritmo correcto es crucial cuando trabajamos con grandes volúmenes de datos.

## Conclusión

- **$O(n)$**: El tiempo crece proporcionalmente al tamaño de los datos.
- **$O(n \log n)$**: Crecimiento moderado, típico de algoritmos de ordenación eficientes.
- **$O(n^2)$**: Crecimiento explosivo, inviable para datos grandes.

Elegir un buen algoritmo puede marcar la diferencia entre segundos y horas de cómputo.
