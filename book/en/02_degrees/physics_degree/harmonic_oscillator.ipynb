# Harmonic Oscillator: Simple Harmonic Motion

Simple harmonic motion describes systems that oscillate periodically:

$x(t) = A\cos(\omega t + \varphi)$

where $A$ is the amplitude, $\omega$ the angular frequency, and $\varphi$ the initial phase.
By modifying these parameters we can explore different oscillatory behaviors.

```python
import numpy as np
import matplotlib.pyplot as plt

A = 2.0          # Amplitude (m)
omega = 2*np.pi  # Angular frequency (rad/s) → 1 Hz
phi = np.pi/4    # Initial phase (rad)

t = np.linspace(0, 4, 1000)
x = A * np.cos(omega * t + phi)

plt.figure(figsize=(10, 4))
plt.plot(t, x, linewidth=2, color='darkorange')
plt.axhline(0, color='gray', linewidth=0.5)
plt.xlabel('Time (s)')
plt.ylabel('Displacement x (m)')
plt.title(f'Harmonic Oscillator  A={A}, ω={omega:.2f}, φ={phi:.2f}')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Effect of Parameters

You can experiment by changing `A`, `omega`, and `phi` in the cell above:

- **Amplitude (A):** controls the maximum height of the oscillation.
- **Angular frequency (ω):** controls the oscillation speed ($T = 2\pi/\omega$).
- **Initial phase (φ):** shifts the curve horizontally.

```python
A = 2.0
phi = 0
t = np.linspace(0, 4, 500)

fig, axes = plt.subplots(1, 3, figsize=(14, 3.5), sharey=True)

for ax, omega_val, color in zip(axes, [np.pi, 2*np.pi, 4*np.pi],
                                  ['steelblue', 'darkorange', 'crimson']):
    x = A * np.cos(omega_val * t + phi)
    ax.plot(t, x, linewidth=2, color=color)
    ax.set_title(f'ω = {omega_val/np.pi:.0f}π rad/s')
    ax.set_xlabel('Time (s)')
    ax.grid(True, alpha=0.3)

axes[0].set_ylabel('x (m)')
plt.suptitle('Frequency comparison', y=1.02)
plt.tight_layout()
plt.show()
```

## Conclusion

The simple harmonic oscillator is the basic model for oscillatory phenomena.
Real systems such as pendulums (for small angles), springs, and LC circuits
follow this behavior. The angular frequency $\omega = \sqrt{k/m}$ connects
the motion to the physical properties of the system.
