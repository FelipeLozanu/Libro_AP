# Drawing Circuits with SchemDraw

In this example we draw a series RC circuit (battery + resistor + capacitor)
using the `schemdraw` library. This type of circuit is fundamental in electronics
and appears in low-pass filters, timers, and coupling circuits.

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing(file='rc_circuit.png') as d:
    d += elm.Battery().up().label('V', loc='bot')
    d += elm.Resistor().right().label('R')
    d += elm.Capacitor().right().label('C')
    d += elm.Line().down()
    d += elm.Line().tox(d._elements[0].start).left()

d.draw()
```

## Circuit Components

- **Battery (V):** provides the electromotive force that drives the current.
- **Resistor (R):** limits current flow and dissipates energy as heat.
- **Capacitor (C):** stores energy as an electric field between its plates.

The time constant of the circuit is $\tau = RC$, which determines the charging
and discharging speed of the capacitor.

```python
R = 1000    # 1 kΩ
C = 1e-6    # 1 µF
tau = R * C
print(f'Time constant τ = RC = {tau*1e3:.2f} ms')
print(f'Time for full charge (~5τ) = {5*tau*1e3:.2f} ms')
```

## Conclusion

With `schemdraw` we can generate clean, professional circuit diagrams.
The time constant $\tau = RC$ is the key parameter: after a time $5\tau$ the
capacitor is considered fully charged (≈99.3% of the source voltage).
