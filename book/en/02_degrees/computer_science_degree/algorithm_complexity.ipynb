# Algorithmic Complexity: Visual Comparison

How does execution time grow as we increase data size?
Let's compare three classic complexities:
- $O(n)$ — linear
- $O(n \log n)$ — linearithmic
- $O(n^2)$ — quadratic

```python
import numpy as np
import matplotlib.pyplot as plt
import time

def measure_time(func, data):
    start = time.perf_counter()
    func(data)
    return time.perf_counter() - start

def linear_search(lst):
    target = lst[-1]
    for x in lst:
        if x == target:
            return x

def sort_data(lst):
    return sorted(lst)

def bubble_sort(lst):
    n = len(lst)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]

sizes = [100, 500, 1000, 2000, 3000, 5000]
t_on, t_onlogn, t_on2 = [], [], []

for n in sizes:
    data = list(range(n, 0, -1))
    t_on.append(measure_time(linear_search, data[:]))
    t_onlogn.append(measure_time(sort_data, data[:]))
    t_on2.append(measure_time(bubble_sort, data[:]))

plt.figure(figsize=(8, 5))
plt.plot(sizes, t_on, 'o-', label='O(n) — Linear search')
plt.plot(sizes, t_onlogn, 's-', label='O(n log n) — Sorting')
plt.plot(sizes, t_on2, '^-', label='O(n²) — Bubble sort')
plt.xlabel('Input size (n)')
plt.ylabel('Time (s)')
plt.title('Algorithmic Complexity Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

## Expected result

The plot clearly shows how $O(n^2)$ grows much faster than $O(n)$ and $O(n \log n)$.
This explains why choosing the right algorithm is crucial when working with large datasets.

## Conclusion

- **$O(n)$**: Time grows proportionally to data size.
- **$O(n \log n)$**: Moderate growth, typical of efficient sorting algorithms.
- **$O(n^2)$**: Explosive growth, unfeasible for large data.

Choosing a good algorithm can mean the difference between seconds and hours of computation.
